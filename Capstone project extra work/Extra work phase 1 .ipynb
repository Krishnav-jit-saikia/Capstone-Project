# Working :

```python
import pandas as pd
import numpy as np
```

# Total Crime :

```python
total_crime = pd.DataFrame({'Population in 2020': population_in_2020 , 'Population in 2023': population_in_2023 , 'Literacy rate 2020': Literacy_rate_2020 , 'Literacy rate 2023': Literacy_rate_2023 , 'Below MPI rate 2020': States_below_MPI_2020, 'Below MPI rate 2023': States_below_MPI_2023 ,'Total crime 2020':Total_crimes_2020 , 'Total crime 2023': Total_crimes_2023})
total_crime
```

```python
total_crime['Growth/Decrease in Population(%)'] = np.where(total_crime['Population in 2020'] == 0, 0, 
                                                     round(((total_crime['Population in 2023'] - total_crime['Population in 2020']) / total_crime['Population in 2020']) * 100, 2))

total_crime['Growth/Decrease in Literacy rate(%)'] = np.where(total_crime['Literacy rate 2020'] == 0, 0, 
                                                     round(((total_crime['Literacy rate 2023'] - total_crime['Literacy rate 2020']) / total_crime['Literacy rate 2020']) * 100, 2))


total_crime['Growth/Decrease in MPI rate(%)'] = np.where(total_crime['Below MPI rate 2020'] == 0, 0, 
                                                     round(((total_crime['Below MPI rate 2023'] - total_crime['Below MPI rate 2020']) / total_crime['Below MPI rate 2020']) * 100, 2))


total_crime['Growth/Decrease in total crime(%)'] = np.where(total_crime['Total crime 2020'] == 0, 0, 
                                                     round(((total_crime['Total crime 2023'] - total_crime['Total crime 2020']) / total_crime['Total crime 2020']) * 100, 2))
```

```python
total_crime
```

```python
total_crime['State'] = state
total_crime
```

```python
last_col_name = total_crime.columns[-1]  # Access the last element of the columns list

# Reorder the columns by placing the last column at the beginning
new_column_order = [last_col_name] + list(total_crime.columns[:-1])  # Concatenate with remaining columns

total_crime = total_crime[new_column_order]
```

```python
total_crime
```

```python
filename = 'Total_crime_2020_2023.csv'
total_crime.to_csv(filename , index = False)
```
