# Analysing the 'Total Crime', 'Crime against women' and 'Crime against children' from 2020 to 2023

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Analysis on Total crime :

```python
df = pd.read_csv('Total_crime_2020_2023.csv')
df
```

```python
plt.figure(figsize = (20,10))
sns.barplot(data = df , x = 'State' , y = 'Growth/Decrease in total crime(%)')
plt.xticks(rotation = 45)
plt.show()
```

```python
# So , the growth of crime can be seen in Arunachal pradesh the most , then Nagaland , Bihar  and Sikkim 
```

```python
sns.scatterplot(data = df,x = 'Growth/Decrease in total crime(%)' , y = 'Growth/Decrease in Population(%)')
plt.show()
```

```python
sns.scatterplot(data = df,x = 'Growth/Decrease in total crime(%)' , y = 'Growth/Decrease in Literacy rate(%)')
plt.show()
```

```python
sns.scatterplot(data = df,x = 'Growth/Decrease in total crime(%)' , y = 'Growth/Decrease in MPI rate(%)')
plt.show()
```

```python
# So , these factors like 'Growth/Decrease in Population(%)' , 'Growth/Decrease in Literacy rate(%)' and 'Growth/Decrease in MPI rate(%)'
# is not  effecting that much here in  'Growth/Decrease in Total  crime(%)'.
```
